```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from mpl_toolkits.mplot3d import Axes3D
```

# **(1) Create the image of a paraboloid with one axis scaled (like an oval paraboloid).**

```python
ang = np.arange(0, 2*np.pi, 0.001)
eixo_z = np.arange(0, 1, 0.001)
ang, eixo_z = np.meshgrid(ang, eixo_z)
r = np.sqrt(eixo_z*2)
eixo_x = r * np.cos(ang)
eixo_y = r * np.sin(ang)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(eixo_x, eixo_y, eixo_z, color='k', alpha=0.4)
plt.show()

twoD = plt.contourf(eixo_x, eixo_y, eixo_z)
plt.show()
```

# **(2) Create the image of a rotated sin using rotation of coordinates.**

```python
eixo_x = np.arange(-4, 4, 0.01)
eixo_y = np.arange(-4, 4, 0.01)
eixo_x, eixo_y = np.meshgrid(eixo_x, eixo_y)
eixo_z = np.sin(eixo_x)
eixo_z = sp.ndimage.rotate(eixo_z, 90, reshape = False) #rotated
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(eixo_x, eixo_y, eixo_z, color='k', alpha=0.4)
plt.show()

twoD = plt.contourf(eixo_x, eixo_y, eixo_z)
plt.show()
```

# **(3) Create the image of a gaussian.**

```python
eixo_x = np.arange(-4, 4, 0.01)
eixo_y = np.arange(-4, 4, 0.01)
eixo_x, eixo_y = np.meshgrid(eixo_x, eixo_y)
eixo_z = np.exp(- 1 * eixo_x**2 - 1 * eixo_y**2)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(eixo_x, eixo_y, eixo_z, color='black', alpha=0.4)
plt.show()

twoD = plt.contourf(eixo_x, eixo_y, eixo_z)
plt.show()
```

# **Crie uma função que gere a imagem de um Gaussiano rotacionado opcionalmente por um ângulo theta e com mx, my, sx, sy como argumentos de entrada.**

```python
theta = 60
mx = 0.5
my = 0.5
sx = 2
sy = 2

eixo_x = np.arange(-5, 5, 0.01)
eixo_y = np.arange(-5, 5, 0.01)
eixo_x, eixo_y = np.meshgrid(eixo_x, eixo_y)
t = np.radians(theta)
sx *= sx
sy *= sy

v1 = np.cos(t)**2 / (2 * sx) + np.sin(t)**2 / (2 * sy)
v2 = np.sin(t)**2 / (2 * sx) + np.cos(t)**2 / (2 * sy)
v3 = np.sin(2 * t) / (4 * sx) - np.sin(2 * t) / (4 * sy)
eixo_z = np.exp(-v1 * (eixo_x - mx)**2 - v2 * (eixo_y - my)**2 - 2 * v3 * (eixo_x - mx) * (eixo_y - my))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(eixo_x, eixo_y, eixo_z, color='black', alpha=0.4)
plt.show()

twoD = plt.contourf(eixo_x, eixo_y, eixo_z)
plt.show()
```
