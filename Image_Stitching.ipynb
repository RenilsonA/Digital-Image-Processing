```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
path = 'drive/MyDrive/'

img1 = cv2.imread(os.path.join(path, 'ic1.jpg'))
img2 = cv2.imread(os.path.join(path, 'ic2.jpg'))
img3 = cv2.imread(os.path.join(path, 'ic3.jpg'))
img4 = cv2.imread(os.path.join(path, 'ic4.jpg'))

stitchy = cv2.Stitcher.create()
(dummy, output)=stitchy.stitch([img1, img2, img3, img4])

plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
plt.show()
```
